```python
genes = dados[dados['tratamento'] == 'com_droga'].loc[:, 'g0':'g49']
cel = dados[dados['tratamento'] == 'com_droga'].loc[:,'c0':'c49']

tabela = pd.concat([genes,cel], axis=1)
tabela
```

```python
corr = tabela.corr().round(2)
corr
```

```python
corr_filtrado = corr.loc[:'g49','c0':]
corr_filtrado.shape
```

```python
mask = np.triu(np.ones_like(corr_filtrado, dtype=bool))
#a variável mask é um triangulo superior (np.triu) de uma array das mesmas dimensões que nossa matrix de correlação
#e seus valores são do tipo booleano e iguais a 1 (True)
f, ax = plt.subplots(figsize=(11, 9)) #nesta etapa é selecionada a área que o gráfico irá ocupar
cmap = sns.diverging_palette(230, 20, as_cmap=True) #nesta etapa é selecionada a paleta de cores para a criação da diferença de cor, indo da por 230 até a 20
sns.heatmap(corr_filtrado, mask=mask, cmap=cmap, center = 0, linewidths=.5, cbar_kws={'shrink': .5}) #neste caso é construído o mapa de calor entre as relações.
```

```python
#Cálculo da correlação entre [g0:g50] e [c0:c50] e plot na forma de heatmap:

sns.set()

cols_genes = [f'g{i}' for i in range(0, 51)]
cols_cel = [f'c{i}' for i in range(0, 51)]
corr_mat = dados[cols_genes+cols_cel].corr().loc[cols_genes,cols_cel]

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(10, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_mat, cmap=cmap, center=0, vmax=1, vmin=-1, square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()
```
