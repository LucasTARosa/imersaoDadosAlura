```python
## esta etapa foram importadas tanto o modelo de machine learning, quanto o modelo de
## teste a ser utilizado para verificar a qualidade do nosso ajuste.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

## nesta etapa são selecionados os dados para alimentar o modelo de machine learning
## neste caso, o x se torna nossas variáveis a serem analisadas, os parâmetros que a
## a maquina deve levar em consideração para retornar o valor sim ou não, que é justamente
## o parâmetro y, que determina a nossa resposta para ela
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

## esta etapa especifica o treinamento e teste do modelo de machine learning, onde foi
## especificado que o volume de treinamento é de 20% (0,2) em relação ao volume de dados 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2)

## aqui é codado o machine learning, com a escolha do modelo a ser utilizado 
## (LogisticRegression), como fazer (fit(x, y)) e o retorno da qualidade do ajuste
## baseado nos testes, que é o (score(x, y)).
modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

após a obtenção do score, é necessário entender se este valor é bom ou ruim. Então a próxima etapa é comparar com outros modelos mais simples, o que é realizado na etapa a seguir

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score


## Este modelo é mais simples do que o anterior, onde leva em consideração que tudo
## é verdadeiro, sendo mais "ingênuo" no tratamento dos dados. 
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

é possível perceber que o primeiro modelo não foi tão superior em relação ao segundo, que assume que todos os valores são verdadeiros. Desta forma, ele não se adequa muito bem ao que desejamos

```python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

## O valor de score dos modelos acaba variando conforme é "rodado" o modelo de machine
## learning, o que se torna um problema quando se deseja comparar diferentes modelos. 
## então de forma a minimizar este efeito, é colocado um parâmetro chamado random_state, 
## com um valor "aleatório", para que se saiba que os valores utilizados entre teste e treino
## sejam os mesmos 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state = 7439)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

Desta forma se garante que que os valores de score obtidos são sempre os mesmos a cada vez que se realizam os testes de machine learning, facilitando para que se realizem comparações entre os modelos

```python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

## Nesta próxima etapa é selecionada a coluna em que há a separação na coluna de dados, 
## tendo uma separação entre as proporções verdadeiras e falsas, no nosso caso o y
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 7439)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 7439)

teste = []
treino = []
for i in range(1, 15):
  modelo_arvore = DecisionTreeClassifier(max_depth = i)
  modelo_arvore.fit(x_treino, y_treino)
  teste.append(modelo_arvore.score(x_teste, y_teste))
  treino.append(modelo_arvore.score(x_treino, y_treino))
```

Neste caso, a medida em que aprofundamos na árvore de treinamento, nosso modelo se torna cada vez melhor em trabalhar com os dados de treino. Entretanto, essa acurácia cada vez maior de treinamento, acaba fazendo com que o modelo diminua sua capacidade de aplicação em dados que ele nunca viu (nosso teste). Este efeito é chamado de overfit

Uma das alternativas para que se consiga superar o overfit no modelo de árvores é randomizando os dados em que as árvores são geradas, criando várias árvores de decisões, utilizando dados aleatórios. O modelo que cria árvores aleatórias é chamado de random_forest, testado a seguir.
