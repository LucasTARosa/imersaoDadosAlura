```python
## esta etapa foram importadas tanto o modelo de machine learning, quanto o modelo de
## teste a ser utilizado para verificar a qualidade do nosso ajuste.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

## nesta etapa sÃ£o selecionados os dados para alimentar o modelo de machine learning
## neste caso, o x se torna nossas variÃ¡veis a serem analisadas, os parÃ¢metros que a
## a maquina deve levar em consideraÃ§Ã£o para retornar o valor sim ou nÃ£o, que Ã© justamente
## o parÃ¢metro y, que determina a nossa resposta para ela
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

## esta etapa especifica o treinamento e teste do modelo de machine learning, onde foi
## especificado que o volume de treinamento Ã© de 20% (0,2) em relaÃ§Ã£o ao volume de dados 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2)

## aqui Ã© codado o machine learning, com a escolha do modelo a ser utilizado 
## (LogisticRegression), como fazer (fit(x, y)) e o retorno da qualidade do ajuste
## baseado nos testes, que Ã© o (score(x, y)).
modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

apÃ³s a obtenÃ§Ã£o do score, Ã© necessÃ¡rio entender se este valor Ã© bom ou ruim. EntÃ£o a prÃ³xima etapa Ã© comparar com outros modelos mais simples, o que Ã© realizado na etapa a seguir

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score


## Este modelo Ã© mais simples do que o anterior, onde leva em consideraÃ§Ã£o que tudo
## Ã© verdadeiro, sendo mais "ingÃªnuo" no tratamento dos dados. 
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

Ã© possÃ­vel perceber que o primeiro modelo nÃ£o foi tÃ£o superior em relaÃ§Ã£o ao segundo, que assume que todos os valores sÃ£o verdadeiros. Desta forma, ele nÃ£o se adequa muito bem ao que desejamos

```python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

## O valor de score dos modelos acaba variando conforme Ã© "rodado" o modelo de machine
## learning, o que se torna um problema quando se deseja comparar diferentes modelos. 
## entÃ£o de forma a minimizar este efeito, Ã© colocado um parÃ¢metro chamado random_state, 
## com um valor "aleatÃ³rio", para que se saiba que os valores utilizados entre teste e treino
## sejam os mesmos 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, random_state = 7439)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

Desta forma se garante que que os valores de score obtidos sÃ£o sempre os mesmos a cada vez que se realizam os testes de machine learning, facilitando para que se realizem comparaÃ§Ãµes entre os modelos

```python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

## Nesta prÃ³xima etapa Ã© selecionada a coluna em que hÃ¡ a separaÃ§Ã£o na coluna de dados, 
## tendo uma separaÃ§Ã£o entre as proporÃ§Ãµes verdadeiras e falsas, no nosso caso o y
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 7439)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```
